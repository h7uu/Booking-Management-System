```python


def calculate_mean(service_data):
    total = 0
    count = 0
    for entry in service_data:
        total += sum(entry[2])
        count += len(entry[2])
    return total / count

def most_crowded_day(service_data):
  max_total = -1
  day_name = ""

  for entry in service_data :
    total = sum(entry[2])
    if total > max_total :
      max_total = total
      day_name = entry[1]
  return day_name , max_total

def least_crowded_day(service_data):
  min_total = float('inf')
  day_name = ""

  for entry in service_data :
    total = sum(entry[2])
    if total < min_total :
      min_total = total
      day_name = entry[1]
  return day_name , min_total

print("The mean of New Car ID service is",round((calculate_mean(new_car_id))))
day_name, max_total = most_crowded_day(new_car_id)
print("The most crowded day in new car ID service is", day_name , "with", max_total , "entries.")
day_name , min_total = least_crowded_day(new_car_id)
print("The least crowded day in new car ID service is", day_name, "with", min_total, "entries.")
```
